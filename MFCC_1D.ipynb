```python
import numpy as np
import os
from os import listdir
import librosa as lb
import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
files_loc = 'split/'
```

```python
diagnosis_csv = "respiratory_sound_database/Respiratory_Sound_Database/patient_diagnosis.csv"
diagnosis = pd.read_csv(diagnosis_csv, names=['pId', 'diagnosis'])
diagnosis.head()
```

```python
categories = diagnosis['diagnosis'].unique()
categories
```

```python
img_train = []
label_train = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'train/'):
  if f == 'Asthma' or f == 'LRTI':
    print("==========================")
  else:
    for i in os.listdir(files_loc+'train/'+f):
      filename = files_loc+'train/'+f+'/'+i
      y,sr = lb.load(filename)
      mfcc = np.mean(lb.feature.mfcc(y=y,sr=sr,n_mfcc=40), axis=1)
      img_train.append(mfcc)
      label_train.append(to_hot_one[f])
      print(f)
      print(i)

```

```python
img_val = []
label_val = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'val/'):
  if f == 'Asthma' or f == 'LRTI':
    print("==========================")
  else:
    for i in os.listdir(files_loc+'val/'+f):
      filename = files_loc+'val/'+f+'/'+i
      y,sr = lb.load(filename)
      mfcc = np.mean(lb.feature.mfcc(y=y,sr=sr,n_mfcc=40),axis=1)
      img_val.append(mfcc)
      label_val.append(to_hot_one[f])
      print(f)
      print(i)

```

```python
img_train = np.array(img_train)
label_train = np.array(label_train)

img_val = np.array(img_val)
label_val = np.array(label_val)
```

```python
print(img_train.shape)
print(img_val.shape)
```

```python
np.save("img_train_40.npy",img_train)
np.save("label_train_40.npy",label_train)

np.save("img_val_40.npy",img_val)
np.save("label_val_40.npy",label_val)
```

```python
counter = Counter(label_train)
print(counter)
```

```python
counter = Counter(label_val)
print(counter)
```

```python
oversample = SMOTE()
img_train_smote, label_train_smote = oversample.fit_resample(img_train, label_train)
```

```python
oversample = SMOTE()
img_val_smote, label_val_smote = oversample.fit_resample(img_val, label_val)
```

```python
counter = Counter(label_val_smote)
print(counter)
```
